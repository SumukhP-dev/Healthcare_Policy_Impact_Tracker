```python
import pandas as pd
import os
```

```python
dataset_names = ["uncleaned mortality", "cleaned mortality",
                 "uncleaned infant mortality", "cleaned infant mortality",
                 "uncleaned racial equity", "cleaned racial equity",
                 "uncleaned county organized health system", "cleaned county organized health system"]
total_data_points = []
fileSizes = []
```

```python
# Load Uncleaned Mortality Dataset
uncleaned_mortality_datatset = pd.read_csv('/content/2016-2023-imi-results-long-view.csv')

# Load Cleaned Mortality Dataset
cleaned_mortality_datatset = pd.read_csv('/content/cleaned_2016-2023-imi-results-long-view.csv')
```

```python
rows, columns = uncleaned_mortality_datatset.shape
print(f"Rows: {rows}, Columns: {columns} Total Data Points: {rows * columns}")
total_data_points.append(rows * columns)

file_path = '/content/2016-2023-imi-results-long-view.csv'

file_size = os.path.getsize(file_path)
print(f"The file size is: {file_size} bytes")
fileSizes.append(file_size)

uncleaned_mortality_datatset.describe(include='all')

columns_to_check = ['Risk Adjuested Mortality Rate', '# of Deaths', '# of Cases', 'LONGITUDE', 'LATITUDE']
for col in columns_to_check:
    uncleaned_mortality_datatset = uncleaned_mortality_datatset[
        (uncleaned_mortality_datatset[col] != ".") &
        (uncleaned_mortality_datatset[col].astype(str) != "NaN") &
        (uncleaned_mortality_datatset[col].notna())
    ]


uncleaned_mortality_datatset["Risk Adjuested Mortality Rate"] = pd.to_numeric(uncleaned_mortality_datatset["Risk Adjuested Mortality Rate"])
uncleaned_mortality_datatset["# of Deaths"] = pd.to_numeric(uncleaned_mortality_datatset["# of Deaths"])
uncleaned_mortality_datatset["# of Cases"] = pd.to_numeric(uncleaned_mortality_datatset["# of Cases"])
uncleaned_mortality_datatset["LONGITUDE"] = pd.to_numeric(uncleaned_mortality_datatset["LONGITUDE"])
uncleaned_mortality_datatset["LATITUDE"] = pd.to_numeric(uncleaned_mortality_datatset["LATITUDE"])

print(uncleaned_mortality_datatset.describe())
```

```python
rows, columns = cleaned_mortality_datatset.shape
print(f"Rows: {rows}, Columns: {columns} Total Data Points: {rows * columns}")
total_data_points.append(rows * columns)

file_path = '/content/cleaned_2016-2023-imi-results-long-view.csv'

file_size = os.path.getsize(file_path)
print(f"The file size is: {file_size} bytes")
fileSizes.append(file_size)

cleaned_mortality_datatset.describe()
```

```python
# Load Uncleaned County Organized Health System Dataset
uncleaned_county_organized_health_system_datatset = pd.read_csv('/content/4.7-cohs-plans-q4-2024-odp.csv')

# Load Cleaned County Organized Health System Dataset
cleaned_county_organized_health_system_datatset = pd.read_csv('/content/cleaned_4.7-cohs-plans-q4-2024-odp (2).csv')
```

```python
rows, columns = uncleaned_county_organized_health_system_datatset.shape
print(f"Rows: {rows}, Columns: {columns} Total Data Points: {rows * columns}")
total_data_points.append(rows * columns)

file_path = '/content/4.7-cohs-plans-q4-2024-odp.csv'

file_size = os.path.getsize(file_path)
print(f"The file size is: {file_size} bytes")
fileSizes.append(file_size)

uncleaned_county_organized_health_system_datatset[" Number of Eligible Individuals "] = pd.to_numeric(uncleaned_county_organized_health_system_datatset[" Number of Eligible Individuals "].apply(lambda x: x.strip().replace(",", "")))

uncleaned_county_organized_health_system_datatset.describe()
```
